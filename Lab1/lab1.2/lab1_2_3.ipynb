```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import glob
%matplotlib inline

#Lab1.2.3
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

catAngry = cv2.cvtColor(cv2.imread('cat1.jpg'),cv2.COLOR_BGR2RGB)
angryCat = cv2.resize(catAngry,(300,300))
newMask = np.zeros(angryCat.shape[:2], dtype="uint8")

cv2.rectangle(newMask, (100, 75), (250, 225),(255,255,255),-1)

res = cv2.bitwise_and(angryCat,angryCat, mask=newMask)

fig, axes = plt.subplots(1, 3)
axes[0].imshow(angryCat)
axes[0].set_title('Original')
axes[1].imshow(newMask)
axes[1].set_title('mask')
axes[2].imshow(res)
axes[2].set_title('Bitwise_AND() result')
```
