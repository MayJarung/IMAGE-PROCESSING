```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import glob
%matplotlib inline

#Lab1.2 [1]

img = cv2.imread('shiba.jpg')
imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
rgb_r = imgRGB[:,:,0]
rgb_g = imgRGB[:,:,1]
rgb_b = imgRGB[:,:,2]
imgHSV = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
hsv_h = imgHSV[:,:,0]
hsv_s = imgHSV[:,:,1]
hsv_v = imgHSV[:,:,2]
imgHLS= cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
hls_h = imgHLS[:,:,0]
hls_l = imgHLS[:,:,1]
hls_s = imgHLS[:,:,2]
imgYCrCb = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)
YCrCb_Y = imgYCrCb[:,:,0]
YCrCb_Cr = imgYCrCb[:,:,1]
YCrCb_Cb = imgYCrCb[:,:,2]


fig, axes = plt.subplots(4, 4, figsize=(16, 16))

axes[0,0].imshow(imgRGB)
axes[0,0].set_title('RGB')

axes[0,1].imshow(rgb_r, cmap='gray')
axes[0,1].set_title('R')

axes[0,2].imshow(rgb_g, cmap='gray')
axes[0,2].set_title('G')

axes[0,3].imshow(rgb_b, cmap='gray')
axes[0,3].set_title('B')

axes[1,0].imshow(imgHSV)
axes[1,0].set_title('HSV')

axes[1,1].imshow(hsv_h, cmap='gray')
axes[1,1].set_title('H')

axes[1,2].imshow(hsv_s, cmap='gray')
axes[1,2].set_title('S')

axes[1,3].imshow(hsv_v, cmap='gray')
axes[1,3].set_title('V')

axes[2,0].imshow(imgHLS)
axes[2,0].set_title('HLS')

axes[2,1].imshow(hls_h, cmap='gray')
axes[2,1].set_title('H')

axes[2,2].imshow(hls_l, cmap='gray')
axes[2,2].set_title('L')

axes[2,3].imshow(hls_s, cmap='gray')
axes[2,3].set_title('S')

axes[3,0].imshow(imgYCrCb)
axes[3,0].set_title('YCrCb')

axes[3,1].imshow(YCrCb_Y, cmap='gray')
axes[3,1].set_title('Y')

axes[3,2].imshow(YCrCb_Cr, cmap='gray')
axes[3,2].set_title('Cr')

axes[3,3].imshow(YCrCb_Cb, cmap='gray')
axes[3,3].set_title('Cb')

```

```python
#Lab1.2.2
catAngry = cv2.imread('cat1.jpg')
cat = cv2.imread('cat2.jpg')
angryCat = cv2.resize(catAngry,(600,600))
memeCat = cv2.resize(cat,(600,600))
img_array = []
for i in range(1,21):
    newImg = cv2.addWeighted(angryCat,0.05*i,memeCat,1-(i*(0.05)),0)
    img_array.append(newImg)
    
output = cv2.VideoWriter('lab1_2_2.mp4',cv2.VideoWriter_fourcc(*'MP4V'),5, (600,600))
for i in range(len(img_array)):
    output.write(img_array[i])
output.release()
```

```python
#Lab1.2.3
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

catAngry = cv2.cvtColor(cv2.imread('cat1.jpg'),cv2.COLOR_BGR2RGB)
angryCat = cv2.resize(catAngry,(300,300))
mask = np.zeros(angryCat.shape[:2], dtype="uint8")

cv2.rectangle(mask, (100, 75), (250, 225),255,-1)

res = cv2.bitwise_and(angryCat,angryCat, mask=mask)

fig, axes = plt.subplots(1, 3)
axes[0].imshow(angryCat)
axes[0].set_title('Original')
axes[1].imshow(mask)
axes[1].set_title('mask')
axes[2].imshow(res)
axes[2].set_title('Bitwise_AND() result')
```
