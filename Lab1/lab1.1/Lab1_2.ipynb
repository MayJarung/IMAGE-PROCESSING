```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
#Lab1.1 (1.1.2)

img_bgr = cv2.imread('mountain.jpg')
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

img_r = img_rgb[:,:,0]

tranposeImg = np.transpose(img_rgb)
tranposeImg_r = tranposeImg[0,:,:]

moveAxisImg = np.moveaxis(img_rgb, -1, 0)
moveAxisImg_r = moveAxisImg[0,:,:]

reShape = np.reshape(img_rgb,(3,667,1000))
reShape_r = reShape[0,:,:]

fig, axes = plt.subplots(1, 5, figsize=(18, 5))

axes[0].imshow(img_rgb)
axes[0].set_title(img_rgb.shape)

axes[1].imshow(img_r)
axes[1].set_title(img_rgb.shape)

axes[2].imshow(tranposeImg_r)
axes[2].set_title(tranposeImg.shape)

axes[3].imshow(moveAxisImg_r)
axes[3].set_title(moveAxisImg.shape)

axes[4].imshow(reShape_r)
axes[4].set_title(reShape.shape)
```
