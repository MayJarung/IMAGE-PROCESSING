```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
#Lab1.1 (1.1.1 manual Slicing)

# Load the image using OpenCV
image_bgr = cv2.imread('Goldfish.jpg')

b = image_bgr[:,:,0]
g = image_bgr[:,:,1]
r = image_bgr[:,:,2]

# Convert the BGR image to RGB format
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

cr = image_rgb[:,:,0]
cg= image_rgb[:,:,1]
cb = image_rgb[:,:,2]



# Create a 2x4 subplot grid for displaying the RGB channels
fig, axes = plt.subplots(2, 4, figsize=(15, 6))

#display BGR picture
axes[0,0].imshow(image_bgr)
axes[0,0].set_title('BGR Picture')

# Display the Blue channel
axes[0,1].imshow(b, cmap='gray')
axes[0,1].set_title('Blue Channel')

# Display the Green channel
axes[0,2].imshow(g, cmap='gray')
axes[0,2].set_title('Green Channel')

# Display the Red channel
axes[0,3].imshow(r, cmap='gray')
axes[0,3].set_title('Red Channel')

#display RGB picture
axes[1,0].imshow(image_rgb)
axes[1,0].set_title('RGB Picture')

# Display the Red channel
axes[1,1].imshow(cr, cmap='gray')
axes[1,1].set_title('Red Channel')

# Display the Green channel
axes[1,2].imshow(cg, cmap='gray')
axes[1,2].set_title('Green Channel')

# Display the Blue channel
axes[1,3].imshow(cb, cmap='gray')
axes[1,3].set_title('Blue Channel')

plt.show()
```
