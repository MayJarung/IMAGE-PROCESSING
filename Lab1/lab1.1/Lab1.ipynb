```python
#Lab1.1 (1.1.1 Spilt Slicing)
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D

# Load the image using OpenCV
image_bgr = cv2.imread('Goldfish.jpg')

# Convert the BGR image to RGB format
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

# Split the RGB image into its color channels
r, g, b = cv2.split(image_rgb)

# Create a 1x3 subplot grid for displaying the RGB channels
fig, axes = plt.subplots(2, 4, figsize=(15, 6))

#display BGR picture
axes[0,0].imshow(image_bgr)
axes[0,0].set_title('BGR Picture')

# Display the Blue channel
axes[0,1].imshow(b, cmap='gray')
axes[0,1].set_title('Blue Channel')

# Display the Green channel
axes[0,2].imshow(g, cmap='gray')
axes[0,2].set_title('Green Channel')

# Display the Red channel
axes[0,3].imshow(r, cmap='gray')
axes[0,3].set_title('Red Channel')

#display RGB picture
axes[1,0].imshow(image_rgb)
axes[1,0].set_title('BGR Picture')

# Display the Red channel
axes[1,1].imshow(r, cmap='gray')
axes[1,1].set_title('Red Channel')

# Display the Green channel
axes[1,2].imshow(g, cmap='gray')
axes[1,2].set_title('Green Channel')

# Display the Blue channel
axes[1,3].imshow(b, cmap='gray')
axes[1,3].set_title('Blue Channel')

plt.show()
```

```python
#Lab1.1 (1.1.1 manual Slicing)

# Load the image using OpenCV
image_bgr = cv2.imread('Goldfish.jpg')

b = image_bgr[:,:,0]
g = image_bgr[:,:,1]
r = image_bgr[:,:,2]

# Convert the BGR image to RGB format
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

cr = image_rgb[:,:,0]
cg= image_rgb[:,:,1]
cb = image_rgb[:,:,2]



# Create a 2x4 subplot grid for displaying the RGB channels
fig, axes = plt.subplots(2, 4, figsize=(15, 6))

#display BGR picture
axes[0,0].imshow(image_bgr)
axes[0,0].set_title('BGR Picture')

# Display the Blue channel
axes[0,1].imshow(b, cmap='gray')
axes[0,1].set_title('Blue Channel')

# Display the Green channel
axes[0,2].imshow(g, cmap='gray')
axes[0,2].set_title('Green Channel')

# Display the Red channel
axes[0,3].imshow(r, cmap='gray')
axes[0,3].set_title('Red Channel')

#display RGB picture
axes[1,0].imshow(image_rgb)
axes[1,0].set_title('RGB Picture')

# Display the Red channel
axes[1,1].imshow(cr, cmap='gray')
axes[1,1].set_title('Red Channel')

# Display the Green channel
axes[1,2].imshow(cg, cmap='gray')
axes[1,2].set_title('Green Channel')

# Display the Blue channel
axes[1,3].imshow(cb, cmap='gray')
axes[1,3].set_title('Blue Channel')

plt.show()
```

```python
#Lab1.1 (1.1.2)

img_bgr = cv2.imread('mountain.jpg')
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

img_r = img_rgb[:,:,0]

tranposeImg = np.transpose(img_rgb)
tranposeImg_r = tranposeImg[0,:,:]

moveAxisImg = np.moveaxis(img_rgb, -1, 0)
moveAxisImg_r = moveAxisImg[0,:,:]

reShape = np.reshape(img_rgb,(3,667,1000))
reShape_r = reShape[0,:,:]

fig, axes = plt.subplots(1, 5, figsize=(18, 5))

axes[0].imshow(img_rgb)
axes[0].set_title(img_rgb.shape)

axes[1].imshow(img_r)
axes[1].set_title(img_rgb.shape)

axes[2].imshow(tranposeImg_r)
axes[2].set_title(tranposeImg.shape)

axes[3].imshow(moveAxisImg_r)
axes[3].set_title(moveAxisImg.shape)

axes[4].imshow(reShape_r)
axes[4].set_title(reShape.shape)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
#Lab1.1 (1.1.3 for method)

#convertRGB to Gray
img_bgr = cv2.imread('catEiei.jpg')
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)

print(img_gray)
print(img_gray[1][1])
quanImg = img_gray.copy()
for i in range(len(img_gray)):
    for j in range(len(img_gray[0])):
        quanImg[i][j] = np.floor((quanImg[i][j]/255)*((2**3)-1))
        
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(img_gray, cmap='gray')
axes[1].imshow(quanImg, cmap ='gray')
```

```python
#Lab1.1 (1.1.4)

#convertRGB to Gray
img_bgr = cv2.imread('bird.jpg')
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)

#ReduceDimension
resize_img_gray = cv2.resize(img_gray,(200,200))

#meshgrid
x,y = np.mgrid[0:200,0:200]

fig3d = plt.figure()
ax = fig3d.add_subplot(projection='3d')
ax.plot_surface(x,y,resize_img_gray,rstride=1,cstride=1,linewidth=0,cmap='gray')




```
